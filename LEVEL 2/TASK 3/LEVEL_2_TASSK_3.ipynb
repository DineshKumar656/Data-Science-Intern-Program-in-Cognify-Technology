##**1: Create New Features from Text Length**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Task 1: Create 'Name Length' and 'Address Length' ---
print("--- Creating text length features ---")

# Calculate the length of the string in 'Restaurant Name'
df['Name Length'] = df['Restaurant Name'].str.len()

# Calculate the length of the string in 'Address'
df['Address Length'] = df['Address'].str.len()

print("Created 'Name Length' and 'Address Length' columns.")
print(df[['Restaurant Name', 'Name Length', 'Address', 'Address Length']].head())
```

##**2: Create a 'Cuisine Count' Feature**

```python
print("\n--- Creating 'Cuisine Count' feature ---")

# --- THIS IS THE FIX ---
# We use str(x) to convert any potential floats (like NaN) to a string *before* splitting
df['Cuisine Count'] = df['Cuisines'].apply(lambda x: len(str(x).split(', ')))

print("Created 'Cuisine Count' column.")
print(df[['Cuisines', 'Cuisine Count']].head())

# Let's check the row that was causing the error
# This will show any rows where 'Cuisines' was not a string
print("\nChecking for non-string 'Cuisines' entries:")
print(df[df['Cuisines'].apply(type) != str])
```

##**3:Confirm Encoding (The "Already Done" Task)**

```python
print("\n--- Confirming 'Has Table booking' and 'Has Online delivery' are encoded ---")

# Check and convert 'Has Table booking'
if pd.api.types.is_object_dtype(df['Has Table booking']):
    print("Converting 'Has Table booking' to 1/0...")
    df['Has Table booking'] = df['Has Table booking'].str.strip().map({'Yes': 1, 'No': 0})
else:
    print("'Has Table booking' is already numeric (1/0).")

# Check and convert 'Has Online delivery'
if pd.api.types.is_object_dtype(df['Has Online delivery']):
    print("Converting 'Has Online delivery' to 1/0...")
    df['Has Online delivery'] = df['Has Online delivery'].str.strip().map({'Yes': 1, 'No': 0})
else:
    print("'Has Online delivery' is already numeric (1/0).")

print("\nEncoding task is confirmed complete.")
```
